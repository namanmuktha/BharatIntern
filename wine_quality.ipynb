```python
import numpy as np

# Define the remove_outliers function
def remove_outliers(col):
    col = sorted(col)  # Sort the values and reassign to col
    Q1, Q3 = np.percentile(col, [25, 75])
    IQR = Q3 - Q1
    lower_range = Q1 - 1.5 * IQR
    upper_range = Q3 + 1.5 * IQR
    return lower_range, upper_range

# Iterate through each column in the DataFrame
for i in wine_data.columns:
    if i!="quality":
           l, r = remove_outliers(wine_data[i])  # Access the column data using wine_data[i]
           wine_data[i] = np.where(wine_data[i] > r, r, wine_data[i])
           wine_data[i] = np.where(wine_data[i] < l, l, wine_data[i])
```

```python
wine_data["sulfur_1"]=wine_data["free sulfur dioxide"]/wine_data["total sulfur dioxide"]
wine_data["sulfur_2"]=wine_data["free sulfur dioxide"]/wine_data["sulphates"]
wine_data["sulfur_3"]=wine_data["total sulfur dioxide"]/wine_data["sulphates"]
wine_data["acidity_1"]=wine_data["citric acid"]/wine_data["volatile acidity"]
wine_data["acidity_2"]=wine_data["citric acid"]/wine_data["fixed acidity"]
wine_data["acidity_3"]=wine_data["fixed acidity"]/wine_data["volatile acidity"]
```

```python
from sklearn.model_selection import train_test_split
wine_train,wine_test=train_test_split(wine_data,test_size=0.2,random_state=13)
X_train=wine_train.drop(["quality"],axis=1)
y_train=wine_train["quality"]
X_test=wine_test.drop(["quality"],axis=1)
y_test=wine_test["quality"]
```

```python
from sklearn.linear_model import  LogisticRegression
log=LogisticRegression()
log.fit(X_train,y_train)
y_pred=log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_pred,y_test)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='gini',max_depth=20,min_samples_split=5)
dt.fit(X_train,y_train)
y_pred=dt.predict(X_test)
print(accuracy_score(y_pred,y_test)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(max_depth=30, n_estimators=300)
rfc.fit(X_train,y_train)
y_pred=rfc.predict(X_test)
print(accuracy_score(y_pred,y_test)*100)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel="rbf",C=1)
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print(accuracy_score(y_pred,y_test)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=6,weights='distance',algorithm="ball_tree",n_jobs=200,p=2)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print(accuracy_score(y_pred,y_test)*100)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,y_train)
y_pred=gnb.predict(X_test)
print(accuracy_score(y_pred,y_test)*100)
```

```python
from sklearn.linear_model import SGDClassifier
sgd=SGDClassifier()
sgd.fit(X_train,y_train)
y_pred=sgd.predict(X_test)
print(accuracy_score(y_pred,y_test)*100)
```

```python
from sklearn.model_selection import GridSearchCV


# Define the hyperparameter grid
param_grid = {
    'alpha': [0.001, 0.01, 0.1],
    'penalty': ['l1', 'l2', 'elasticnet'],
    'loss': ['hinge', 'log'],
    'max_iter': [100, 500, 1000],
}

# Create an SGDClassifier
sgd = SGDClassifier()

# Perform grid search with cross-validation
grid_search = GridSearchCV(sgd, param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
best_model = grid_search.best_estimator_
```
